```python
wifi_data['event_timestamp'] = pd.to_datetime(wifi_data['event_timestamp'])
```

```python
# Extract the hour component from event_timestamp
wifi_data['hour'] = wifi_data['event_timestamp'].dt.hour

# Group by poc_id and hour, and count the number of interactions
interactions_by_hour = wifi_data.groupby(['poc_id', 'hour']).size().reset_index(name='interaction_count')

# Find the POC IDs with the highest interaction counts for each hour
max_interactions_by_hour = interactions_by_hour.groupby('hour').apply(lambda x: x[x['interaction_count'] == x['interaction_count'].max()])

# Display the POC IDs with the highest interaction counts for each hour
print(max_interactions_by_hour)
```

```python
interactions_by_hour = wifi_data.groupby(['poc_id', 'hour']).size().reset_index(name='interaction_count')

# Sort the interactions by count in descending order
interactions_by_hour_sorted = interactions_by_hour.sort_values(by=['hour', 'interaction_count'], ascending=[True, False])

# Define a function to get the top n POC IDs for each hour
def top_n_poc_ids(group, n=10):
    return group.head(n)

# Apply the function to get the top 10 POC IDs for each hour
top_10_poc_ids_by_hour = interactions_by_hour_sorted.groupby('hour').apply(top_n_poc_ids)

# Display the top 10 POC IDs with the highest interaction counts for each hour
print(top_10_poc_ids_by_hour)
```
